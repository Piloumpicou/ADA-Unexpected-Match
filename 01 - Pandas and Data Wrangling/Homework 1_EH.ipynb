## Task 1. Compiling Ebola Data

The `DATA_FOLDER/ebola` folder contains summarized reports of Ebola cases from three countries (Guinea, Liberia and Sierra Leone) during the recent outbreak of the disease in West Africa.
```python
# Write your answer here
guinea_path = DATA_FOLDER + '/ebola/guinea_data'
guinea_files = glob.glob(os.path.join(guinea_path, "*.csv"))
data_guinea = pd.concat((pd.read_csv(file)) for file in guinea_files)
data_guinea = data_guinea.reset_index(drop=True)

columns=['Date','Description','Totals',
         'Conakry','Gueckedou','Macenta',
         'Dabola','Kissidougou','Dinguiraye',
         'Telimele','Boffa','Kouroussa',
         'Siguiri','Pita','Mzerekore',
         'Yomou','Dubreka','Forecariah',
         'Kerouane','Coyah','Dalaba',
         'Beyla','Kindia','Lola']

#reorders the columns according to the order specified in 'columns'
data_guinea = data_guinea[columns]
```

```python
liberia_path = DATA_FOLDER + '/ebola/liberia_data'
liberia_files = glob.glob(os.path.join(liberia_path, "*.csv"))
data_liberia = pd.concat((pd.read_csv(file, )) for file in liberia_files)
data_liberia = data_liberia.reset_index(drop=True)

columns=['Date','Variable','National',
         'Bomi County','Bong County','Grand Kru',
         'Lofa County','Margibi County','Maryland County',
         'Montserrado County','Nimba County','River Gee County',
         'RiverCess County','Sinoe County']

#reorders the columns according to the order specified in 'columns'
data_liberia = data_liberia[columns]
```

```python
data = data_guinea.copy()
data.Description[data.Description == 'New deaths registered today'] = 'New deaths registered'
data = data.loc[(data['Description'] == 'New cases of suspects')
              | (data['Description'] == 'New cases of probables')
              | (data['Description'] == 'New cases of confirmed')
              | (data['Description'] == "New deaths registered")]

data['Total Guinea'] = pd.to_numeric(data['Total Guinea'])
data = data[np.isfinite(data['Total Guinea'])]
```

```python
#convert the 'Date' to a Date of the form yyyy-mm-dd
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%d')
#convert the 'Total Guinea' to an int32
#data['Total Guinea'] = data['Total Guinea'].astype(str).astype(int)
#data = data[np.isfinite(data['Total Guinea'])]
#set 'Date' as an index in order to be able to groupby month
data.set_index(data['Date'], inplace=True)
data.groupby([pd.TimeGrouper(freq='M'), 'Description'])['Total Guinea'].sum()
```

```python
#reset the 'Date' index to the default integer index
data.reset_index(drop=True, inplace=True)
```
